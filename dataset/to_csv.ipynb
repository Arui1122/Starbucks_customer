### 因為 value 欄位內是 dict 形式因此將 value 切分成三個欄位 offer id, reward , amount

```python
import json

def offerId(X):
    if 'offer id' in X:
        return X['offer id']
    elif 'offer_id' in X:
        return X['offer_id']
    else:
        return None
    
def reward(X):
    if 'reward' in X:
        return X['reward']
    else:
        return None
    
def amount(X):
    if 'amount' in X:
        return X['amount']
    else:
        return None
```

```python
p['offer_id']=p['value'].apply(offerId)
p['reward']=p['value'].apply(reward)
p['amount']=p['value'].apply(amount)
```

```python
p.tail()
```

```python
p.drop(columns=['value'],inplace=True)
```

```python
p.head()
```

```python
p.to_csv('transcript.csv',index=False)
```

# Deal with portfolio

```python
# ref : https://www.kaggle.com/code/hugovallejo/starbucks-data-random-forest-approach-93-acc
portfolio=pd.read_json('portfolio.json',lines=True)
```

```python
portfolio.head()
```

```python
channels = ["web", "email", "mobile", "social"]
```

```python
def encode_channel(channel, channel_name):
  return 1 if channel_name in channel else 0
```

```python
for channel in channels:
  portfolio[f'{channel}_channel'] = portfolio['channels'].apply(encode_channel, channel_name=channel)
portfolio
```

```python
portfolio.drop(columns=["channels"], inplace=True)
portfolio
```

```python
def encode_offerType(offertype):
  if offertype =='bogo':
      return 0
  elif offertype =='discount':
    return 1
  else:
     return 2
```

```python
portfolio['offer_type']=portfolio['offer_type'].apply(encode_offerType)
```

```python
# portfolio = pd.get_dummies(portfolio, columns=["offer_type"], drop_first=True)
portfolio = portfolio.rename(columns={"id":"offer_id"})
portfolio
```

```python
portfolio.to_csv('portfolio.csv',index=False)
```

# Deal with profile

```python
profile=pd.read_json('profile.json',lines=True)
```

```python
profile.head()
```

```python
profile.info()
```

```python
profile['became_member_on'] = pd.to_datetime(profile['became_member_on'], format='%Y%m%d')
```

```python
print(profile['became_member_on'].min())
print(profile['became_member_on'].max())
print(profile['became_member_on'].max()-profile['became_member_on'].min())
```

```python
# 以 2018 12 31 為基準點
dayNow=time_point = pd.to_datetime("12/31/2018")
def become_member_days(date):
    day=dayNow-date
    return day
```

```python
profile['became_member_days']=profile['became_member_on'].apply(become_member_days)
```

```python
profile['became_member_days']=profile['became_member_days'].dt.days
```

```python
profile.info()
```

```python
profile.dropna(subset=["gender", "income"], inplace=True)
```

```python
profile = pd.get_dummies(profile, columns=["gender"], drop_first=True)
profile
```
