```python
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True)
pd.set_option('display.max_rows', 100000)
pd.set_option('display.max_columns', 100000)
```

```python
transcript=pd.read_csv('./dataset/transcript.csv')
portfolio=pd.read_csv('./dataset/portfolio.csv')
```

```python
# event 有三種形式 offer received , offer completed , offer viewed
transcript.head()
```

```python
portfolio.head()
```

```python
# 將 offer_received 獨立成一個欄位
offer_received=transcript[transcript['event']=='offer received']
offer_received['offer_received']=1
offer_received=offer_received.rename(columns={'time':'received_time'})
offer_received = offer_received.drop(columns=['reward', 'amount','event'])
offer_received.head()
```

```python
# 將 offer_completed 獨立成一個欄位
offer_completed=transcript[transcript['event']=='offer completed']
offer_completed['offer_completed']=1
offer_completed=offer_completed.rename(columns={'time':'complete_time','reward':'complete_reward'})
offer_completed = offer_completed.drop(columns=['amount','event'])
offer_completed.head()
```

```python
# 將 offer_received 和 offer_completed 合併起來
T=pd.merge(offer_received,offer_completed,on=['person','offer_id'] ,how='left')
T.head()
```

```python
# received 都為 1 因此 drop 掉
T = T.drop(columns=['offer_received'])
T.head()
```

```python
# 將每個優惠的特徵合併並進來
T=pd.merge(T,portfolio,on=['offer_id'] ,how='left')
T.head()
```

```python
#針對 discount 優惠卷做分析 , offer type = 1
discount=T[(T['offer_type']==1)]
discount.shape
```

```python
# all web amd email channel = 1 so drop both
discount['offer_completed']=discount['offer_completed'].fillna(0)
discount=discount.drop(columns=['web_channel', 'email_channel'])
discount.head()
```

```python
# 想觀察在discount 優惠卷中，優惠卷的特徵是否會影響消費者購買行為
import statsmodels.formula.api as sm 
y=discount['offer_completed'] # 1 代表有使用優惠卷購買 ， 0 代表沒有使用優惠卷
f='y ~ reward+difficulty+duration'
model=sm.logit(formula=f,data=discount)
result=model.fit()
print(result.summary())
```

```python
# 可以發現 reward 和 duration 都可以正向顯著影響使用優惠卷，因此想探討兩者有沒有存在交互作用
discount['interaction']= discount['reward']*discount['duration']
discount.head()
```

```python
import statsmodels.formula.api as sm 
y=discount['offer_completed']
f='y ~ reward+duration+interaction'
model=sm.logit(formula=f,data=discount)
result=model.fit()
print(result.summary())
```

根據此結果可以看到不論是 reward 或 duration 提升一單位時，會受到另一個變數影響，且兩者呈現負向交互作用，因此兩個變數同時提升一單位時會造成使用優惠卷購買行為下降。</br>
因此當星巴克業者在推動discount優惠卷時，應該選擇調整 reward 或 duration 其中一個即可，避免同時調整。
